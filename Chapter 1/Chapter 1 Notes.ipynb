## The Analytic Approach

### Gallileo:
- horizontal position of ball (x) = (starting speed in horizontal direction) * (time)
- vertical position of ball (height, y) = 

((starting speed in vertical direction) * time) PLUS

((constant downward acceleration due to gravity ~ -9.81 m/s) * time^2) /2


```python
# Pg. 3 (Listing 1-1)

# Appropriate for a hypothetical ball whose initial horizontal speed is about 0.99 m/s
# and whose initial vertical speed is about 9.9 m/s

def ball_trajectory(horiz_speed):
    """A function that calculates the trajectory of a ball."""
    location = 10*horiz_speed - 5*(horiz_speed**2)
    return(location)

#Test

horiz_speed = 10
result = ball_trajectory(horiz_speed=horiz_speed)
print(result)
```

```python
# Pg 3 - List 1.2 - Gravity's Rainbow

import matplotlib.pyplot as plt
xs = [x/100 for x in list(range(201))]
ys = [ball_trajectory(x) for x in xs]
plt.plot(xs, ys)
plt.title('The Trajectory of a Thrown Ball')
plt.xlabel('Horizontal Position of Ball')
plt.ylabel('Vertical Position of Ball')
plt.axhline(y=0)
plt.show()
```

### Tangent Calculation

Chapman observed that when the outfielder is standing in the right location, the tangent of theta grows at a simple, contant rate. (The tangent of thea, the ball's angle with the ground, grows linearly ove time).
- An elegant, algorithmic solution to the outfielder problem: if something grows at a constant rate, it has zero acceleration.
    - If you're standing too close to the ball's initial position, you'll observe positive acceleration
    - If you're standing too far from the ball's initial position, you'll observe negative acceleration
    - So you know where to stand by feeling how steadily you have to tilt back your head 
        - Think with your neck.
